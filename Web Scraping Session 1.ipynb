# Understanding Web Pages #
You have all heard of *HTML* and maybe of *CSS*. We need to know what those are. 

## The Language of the Web ##
The Web is primarily written in HTML — <b>H</b>yper<b>T</b>ext <b>M</b>arkup <b>L</b>anguage. 

In its simple form, it looks like this: 

     <!DOCTYPE html>
     <html lang="en">
       <head>
         <meta charset="utf-8">
         <title>Example page</title>
       </head> 
       <body>  
         <h1>Hello world</h1>
       </body>
    </html>

From here we can distinguish three (3) key elements: 

1. a document type declaration (e.g., `<!DOCTYPE ..>`) which only appears once, always at the top
2. various [HTML Elements](http://docs.webplatform.org/wiki/html/elements) (e.g., `<h1> .. </h1>`)
3. various attributes (e.g. `< .. lang="en">`)

A few points that will be important for our purposes:

* most tags have an opening (`<xx>`) and a closing (`</xx>`), everything in between is the tags 'content'
 * there are exceptions, such as the [br](http://docs.webplatform.org/wiki/html/elements/br) or [img](http://docs.webplatform.org/wiki/html/elements/img) tags which are "self-closing" (written as `<br>` or as `<br />`)
 * the content of a tag can be simply `text` or it can be other tags (html)
* two attributes will be especially useful (`id` and `class`):
 * the `id="xxxx"` attribute refers to _exactly_ one HTML elements (no two elements will have the same id)
 * the `class="xxxx"` is used to assign a [CSS](http://docs.webplatform.org/wiki/css) class, which can be useful for identifying the elements of the page we are interested in


A great list of HTML tutorials and other resources can be found on [WebPlatform.org](http://docs.webplatform.org/wiki/html/tutorials)

Say we want to get the product details. We can start by getting to the right part of the page by finding an `id=` tag. Ideally, we can find a tag that wraps around all the content we want (alternatively, we can find a tag that acts as a marker). 

Look through and try to find `id="productDetails"`

```python
# import some elements we need
import urllib, urllib2 # for fetching pages
from bs4 import BeautifulSoup, Comment # for parsing pages
```

```python
# we use urllib2 to do a page request and read the result
url = 'http://www.amazon.com/Hundred-Solitude-Gabriel-Garcia-Marquez/dp/006112009X/'
content = urllib2.urlopen(url).read()
```

### Congrats! You just scraped your first Web Page! ###
Now lets see a sample of what you've got: 

```python
print content.strip()[0:250] # strip to take out whitespace, 0:250 for first 250 characters only

# note: urllib strips skips everything before the <html> tag
```

Now we need to make some [Beautiful Soup](http://www.crummy.com/software/BeautifulSoup/bs4/doc/)!

From the [docs](http://www.crummy.com/software/BeautifulSoup/bs4/doc/) (which you should read!):
> Beautiful Soup is a Python library for pulling data out of HTML and XML files. It works with your favorite parser to provide idiomatic ways of navigating, searching, and modifying the parse tree. It commonly saves programmers hours or days of work.

```python
# make soup (this parses the page and gives us a structure we can work more easily)
soup = BeautifulSoup(content, "html.parser")
```

Make a python list that contains all the `<li>` tags corresponding to the product details

```python
# remember, we already have the soup based on the page contents
# make soup (this parses the page and gives us a structure we can work more easily)
soup = BeautifulSoup(content, "html.parser")

# hint: you want to first get close to find something near the <li> then you want to find all the <li> 

# answer: 
details = soup.find(id='productDetailsTable').find_all('li')

# save the list in a variable called "details"
```

```python
for d in details:
    print d # a mini-soup of just the HTML inside the <li>
    print d.get_text() # just the text (http://www.crummy.com/software/BeautifulSoup/bs4/doc/#get-text)
    print "--"
```
